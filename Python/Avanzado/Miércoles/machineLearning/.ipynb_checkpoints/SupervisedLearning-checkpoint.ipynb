### K-Nearest Neighbors

Basa su predicción en los 'k' puntos más cercanos a la muestra de la que se quiere conocer su variable objetivo, dependiendo de a que  valor tengan la mayoría de estos puntos más cercanos, será el valor que se va a asignar.

Imaginemos que tenemos un grupo de muestras de las cuales ya conocemos sus respectivos valores, estas son los puntos rojos y verdes.



Y ahora debemos identificar el valor de la nueva muestra, la de color azul.



Si k es igual a 3



El punto tendría el valor de un punto rojo



Si k es igual a 5



El punto tendría el valor de un punto verde



### Todos los modelos de machine learning en Python son implementados como clases

### Las clases implementan los algoritmos para aprendisaje y predicción además de que almacenan los datos usados para aprendisaje

### Al proceso de entrenar un modelo por medio de datos se le conoce como 'fitting' y se realiza por lo general con el método .fit()

### Para predecir nuevos datos, se utiliza el método .predict()

Para utilizar el modelo K-Nearest Neighbors, tenemos que importarlo de sklearn.neighbors

Después instanciamos nuestro clasificador, con parámetro ponemos el valor de k a utilizar

Utilizamos el método fit para entrenar a nuestro clasificador, como primer parámetro colocamos las características y como segundo parámetro los objetivos, ambos como una arreglo de numpy, también pueden ser DataFrames, pero tienen que tener las dimensiones adecuadas. Tanto las características como los objetivos tienen que ser de carácter numério.

```python
clasificador.fit(iris.data, iris.target);
```

Es necesario que las características estén en un arreglo donde cada columna sea una observación y cada columna sea una característica

### Para probar la eficiencia de el modelo que creamos, por lo general, primero se divide los datos que poseemos en dos grupos, un grupo lo vamos a usar para entrenar a nuestro clasificador, el otro lo vamos a utilizar para realizar predicciones con nuestro modelo y comprar los valores reales con los valores que se predijeron.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
caracts = iris.data
objetivos = iris.target
```

```python
caracts_E, caracts_P, obj_E, obj_P = train_test_split(caracts, objetivos, test_size = 0.3, random_state = 30, stratify = objetivos)
```

```python
len(caracts_E), len(caracts_P)
```

```python
len(obj_E), len(obj_P)
```

```python
clasificador = KNeighborsClassifier(n_neighbors = 5)
```

```python
clasificador.fit(caracts_E, obj_E);
```

```python
objPred = clasificador.predict(caracts_P)
```

```python
len(obj_P), len(objPred)
```

### De esta forma podemos medir la exactitud de nuestro modelo

```python
accuracy_score(obj_P, objPred)
```

### Así,  podemos decir que el modelo que nosotros creamos tiene una exactitud de 93%

Existen diferentes formas de optimizar la exactitud de nuestro modelo, por ejemplo, podríamos variar en número de vecinos e identificar cual genera una mayor exactitud.

```python
for i in range(4,8):
    clasificador = KNeighborsClassifier(n_neighbors= i)
    clasificador.fit(caracts_E, obj_E)
    objPred = clasificador.predict(caracts_P)
    print('Para %s vecinos, exactitud : %s' % (i, accuracy_score(obj_P, objPred)))
```

### Podemos realizar una gráfica para ver si el número de habitaciones está correlacionado con el valor de las casas

```python
plt.style.use('ggplot')
plt.scatter(habitaciones,costos)
plt.ylabel("Valor de casas en 1000's $")
plt.xlabel("Número de habitaciones");
# Entre mayor sea el número de habitaciones, mayor es el costo
```

### Regresión lineal

El objetivo de la regresión lineal es ajustar una línea en ciertos datos. 
Una línea en dos dimensiones siempre está en la forma de:
* y = ax + b

Donde:

* y = objetivo
* x = característica única
* a,b = parámetros del modelo

Si por ejemplo contasemos con los siguientes datos


Nuestro objetivo principal es obtener una recta que se aproxime lo más posible a los distintos puntos de datos



De tal forma que la distancia entre la recta y los diferentes puntos, sea la menor posible, en realidad es el cuadrado de las distancias.



Para poder crear un módelo de rergesión lineal, podemos importar el siguiente de sklearn.linear_model

```python
from sklearn.linear_model import LinearRegression
```

Creamos a nuestro regresor

```python
regresor = LinearRegression()
```

Lo entrenamos con los datos con los que queremos modelar

```python
regresor.fit(habitaciones,costos);
```

Con el regresor entrenado, podemos realizar nuevas predicciones

```python
regresor.predict([[7]]) # Costo para casa con 7 habitaciones
```

```python
habitacionesPrueba = np.linspace(min(habitaciones), max(habitaciones),100).reshape(-1,1)
```

```python
habitacionesPrueba
```

```python
costosPredecidos = regresor.predict(habitacionesPrueba)
```

```python
plt.style.use('ggplot')
plt.scatter(habitaciones,costos)
plt.ylabel("Valor de casas en 1000's $")
plt.xlabel("Número de habitaciones");
plt.plot(habitacionesPrueba, costosPredecidos,color = 'black')
```

```python
print('a -> %s'% regresor.coef_)
print('b -> %s'% regresor.intercept_)
```

### Así como la exactitud mide que tan bueno es un modelo de clasificación, R^2 (coeficiente de determinación) nos ayuda a saber que tanto se ajusta nuestro modelo a la variable de interés. Entre más cercano esté a 1, es mejor el ajuste

```python
# R^2
regresor.score(habitaciones,costos)
```
