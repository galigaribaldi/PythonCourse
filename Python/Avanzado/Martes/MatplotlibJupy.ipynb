```python
#Colocar una gráfica dentro de otra gráfica
fig = plt.figure() #fig es una función que nos permite trabajar con add_axes y es de tipo figure
grafica1 = fig.add_axes([0.1,0.1,0.9,0.8]) #son ejes grandes, los definimos
grafica2 = fig.add_axes([0.2,0.5,0.3,0.3]) # los los ejes pequeños, los definimos
#Para agregar color de fondo facecolor='g'
#projection='lambert', o 'polar', son ejemplos de proyecciones

grafica1.plot(x,y,'b')
grafica2.plot(x,y*y,'m')
```

```python
from numpy import random# Importo el módulo de numeros aleatorios de numpy

# se utiliza la función randn() del  random para generar una lista de números aleatorios con distribución normal.
numeros = random.randn(2000)  # array con 200 números aleatorios

plt.title('Histograma con números aleatorios. Distribución normal') # LE doy un título a mi  gráfica
plt.hist(numeros, bins = 20) #Realizo el histograma. bins me indica en cuantas partes voy a dividir el eje x
plt.grid(True) # con grid permito que se muestre una cuadrícula
plt.show() #Para mostrar la gráfica, en Jupyter no es necesario
```

# Gráficas de Barras
## matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center', **kwargs)
 * x es la distancia que va a tener cada dato entre sí
 * heigh s la altura de las barras
 * width es el ancho de las barras, desde 0 hasta 1
 * bottom son la (s) coordenada (s) de las bases de las barras (por defecto: 0).
 * align nos permite centrar o acercar a la derecha o izquierda las barras respecto a los ejes

```python
#Para la comprensión del tema graficaremos als calificaciones de los  5 alumnos de una clase.
alumnos = ["Any","Chuchin", "Diego","José","Natita"] # guardamos en una lista los nombres de los alumnos
calificaciones = [10,5,10,8,4] # guardamos en una lista las calificaciones de los alumnos
xx = range(len(calificaciones)) #puntos en los que s evan a trazar las barras. va desde 0 hasta el numero de elementos dela lista calificaciones
plt.bar(xx,calificaciones,width = 0.5,color = 'r',tick_label = alumnos)
#xx es igual a 5, porque tenemos 5 datos, calificaciones es el siguiente parámetro porque son los datos que queremos que se grafiquen
#color nos permite definirlo rojo, y tick_label nos permite asignarle un nombre a cada barra
```

# Gráfica de pastel
##  matplotlib.pyplot.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=None, radius=None, counterclock=True, wedgeprops=None, textprops=None, center=(0, 0), frame=False, rotatelabels=False, *, data=None)
 
* x: como una matriz.Los tamaños.
* explode: similar a una matriz, opcional, predeterminado: ninguno
* labels: lista, opcional, por defecto: ninguna
* colors: tipo matriz, opcional, por defecto: ninguno
* pctdistance: float, opcional, por defecto: 0.6. La proporción entre el centro de cada sector circular y el inicio del texto generado por autopct.
* shadow: dibuja una sombra debajo del pastel.

```python
# se encuestaron 7 personas para conocer sus pasatiempos
Z = [3,6,7,8,8,10,6]
L = ['comer','jugar','cantar','otra cosa','nadar','leer','dormir']
plt.pie(Z ,labels = L, radius = 1, shadow = True)
```
