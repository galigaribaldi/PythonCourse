# Transformada de Fourier

La transformada de Fourier se realiza con la finalidad de pasar al dominio de la frecuencia señales que originalmente se encuentran en el dominio del tiempo. Se trata de una transformación reversible, es decir, si transformamos una señal en el dominio del tiempo y la pasamos al dominio de la frecuencia, es posible hacer la operación inversa y regresarla al dominio del tiempo. 

Su cálculo con ayuda de SciPy es muy sencillo, y basta con utilizar un vector con los valores de nuestra señal muestreada para calcular su transformada de Fourier y poder hacer los análisis que necesitemos en el dominio de la frecuencia. De igual manera, es posible hacer la transformada inversa de Fourier. Tanto la transformada como la antitransformada se calculan mediante el algoritmo de la transformada rápida de Fourier (fft).

```python
from scipy.fftpack import fft,ifft
from numpy import array

x = array([1.0,2.0,1.0,-1.0,1.5])
#Calculamos la transformada rápida de Fourier con fft
y = fft(x)
print(len(y))
print("Vector original: ",x)
print("Transformada de Fourier: ",y)
xdenuevo = ifft(y) 
print("Señal original: ",xdenuevo)
```

# SciPy Integrate

Cuando una función no se puede integrar de forma analítica o simplemente nos queremos evitar la fatiga (o necesitamos hacerlo), generalmente se recurre a los métodos numéricos de integración. Muchos de ellos se encuentran en la librería scipy.integrate.

```python
import scipy.integrate
from numpy import exp
f= lambda x:exp(-x**2)
#Con quad realizamos integraciones simples
#El primer argumento es la función, y posteriormente el rango de integración
i = scipy.integrate.quad(f, 0, 1)
print(i)
#El resultado regresa dos valores:
#El primero es el resultado de la integración
#El segundo es el error absoluto estimado.
```

Ahora, veremos como resolver una integral doble, por ejemplo:
![](integraldoble.jpg)

```python
from math import sqrt
f = lambda x,y:2*x*y
g = lambda x : sqrt(x)
h = lambda x : 1+x
#La integración doble tiene el siguiente formato:
#scipy.integrate.dblquad(función,limiteinferior1, limitesuperior1, limiteinferior2,limitesuperior2)
i = scipy.integrate.dblquad(f,0,1,g,h)
print(i)
```

Y una integral triple, como: ![](integralTriple.png)

```python
from numpy import cos, sin, Inf, pi
f2 = lambda x,y,z : (x**2)*y*(cos(pi*z)+2)
j = scipy.integrate.tplquad(f2,0,5,0,2,0,3)
print(j)
```

# SciPy Interpolate

Una interpolación es el preceso de encontrar un valor en medio de dos puntos dados. Pueden ser dos puntos de una recta o de una curva. Para ejemplificarlo con Python, haremos el siguiente programa.

```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt

#Obtenemos un vector de 12 elementos con números del 0 al 4
x = np.linspace(0,4,12)
y = np.cos(((x**2)/3)+4)
print( x)
print(y)
```

```python
%matplotlib inline
plt.plot(x,y,'o')
plt.show()
```

```python
#Para hacer una interpolación en una dimensión la clase interp1d es muy útil
#Nos permite crear una función basados en el ajuste de una serie de puntos dados
# Para crear una función interpolada con la información de arriba:
f1 = scipy.interpolate.interp1d(x,y,kind='linear')
f2 = scipy.interpolate.interp1d(x,y,kind='cubic')

print(f1(.464)) #Para una entrada x, devuelve un valor y

xnueva = np.linspace(0,4,30)
plt.plot(x,y,'o',xnueva,f1(xnueva),'-',xnueva,f2(xnueva),'--')
plt.legend(['datos','interp. lineal','interp. cúbica'])
plt.show()
```

Un spline, hablando de matemáticas, es una curva diferenciable definida en porciones mediante polinomios.

![](spline.png)

Es posible usar splines en python para interpolar y representar datos, la clase UnivariateSpline es un método conveniente para crear una función basada en el ajuste de datos brindados.

```python
x = np.linspace(-3, 3, 50)
y = np.exp(-x**2) + 0.1 * np.random.randn(50)
print(len(y))
plt.plot(x, y, 'ro', ms = 5)
plt.show()
```

```python
sp1 = scipy.interpolate.UnivariateSpline(x,y)
x2 =  np.linspace(-3,3,1000)
plt.plot(x2,sp1(x2),'g',lw=3)
plt.show()
```

```python
sp1.set_smoothing_factor(1) #Modificando este argumento, podemos hacer una curva más o menos suavizada
plt.plot(x2, sp1(x2), 'b')
plt.show()
```

# SciPy LinAlg

Este paquete contiene una multitud de rutinas útiles para algebra lineal. La diferencia entre linalg de SciPy y linalg de NumPy es que SciPy contiene algunas funciones avanzadas que no están presentes en numpy.linalg, además de que la versión de SciPy puede llegar a ser más veloz.

```python
#Vamos a solucionar las ecuaciones simultáneas:
# 3x + 2y = 2
# 1x - 1y = 4
# 5y + 1z = -1

from scipy import linalg

#Primero hay que declarar los siguientes arrays con ayuda de numpy
a = np.array([[3,2,0],[1,-1,0],[0,5,1]])
b = np.array([2,4,-1])

#Ahora hay que pasar los valores para resolver
x = linalg.solve(a,b)
print(x)
```

```python
#Ahora vamos a calcular el determinante de una matriz cuadrada
A = np.array([[1,2,13],[3,4,8],[5,2,12]])
x = linalg.det(A)
print(x)
```

```python
#También podemos obtener los eigen-vectores y eigen-valores de una matriz:
B = np.array([[1,2],[3,4]])

l,v = linalg.eig(B)

#Eigenvalores
print(l)
#EigenVectores
print(v)
```
